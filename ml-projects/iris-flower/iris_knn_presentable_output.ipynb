```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split into train & test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)

#predictions
y_pred = clf.predict(X_test)

# Map target numbers to names
target_names = iris.target_names  # ['setosa' 'versicolor' 'virginica']
y_pred_names = [str(target_names[i]) for i in y_pred]
y_test_names = [str(target_names[i]) for i in y_test]

# Display side by side using pandas
import pandas as pd
pd.set_option("display.max_rows", None)

df_compare = pd.DataFrame({
    "Actual Type": y_test_names,
    "Predicted Type": y_pred_names
})

# Print the first 10 predictions to check
print(df_compare)

```
